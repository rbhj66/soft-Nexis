```python
import pandas as pd

# 🔹 Step 1: Load Data
df = pd.read_csv("Supplement_Sales_Weekly_Expanded.csv")

print("Initial Data:")
print(df.head())
print(df.info())

# 🔹 Step 2: Remove Duplicates
df = df.drop_duplicates()

# 🔹 Step 3: Remove Unnecessary Columns (edit if needed)
df = df.drop(columns=["notes", "temp_id"], errors='ignore')

# 🔹 Step 4: Handle Missing Values
print("\nMissing Values Before:")
print(df.isna().sum())

# Fill numerical columns with mean
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill categorical columns with mode
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# 🔹 Step 5: Fix Data Types
for col in df.columns:
    try:
        df[col] = pd.to_datetime(df[col])
    except:
        pass

# 🔹 Step 6: Clean Text Data
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.lower().str.strip()

# 🔹 Step 7: Final Check
print("\nMissing Values After:")
print(df.isna().sum())

print("\nDuplicates Left:", df.duplicated().sum())

# 🔹 Step 8: Save Clean Data
df.to_csv("cleaned_supplement_sales.csv", index=False)

print("\n✅ Cleaning Completed & File Saved!")
```
